**Metrics**:
- відносна кількість правильно суміщених ознак

- похибка локалізації (відстань між реальним розміщенням предмета в кадрі та розпізнаним)

- відносний час обробки фото в залежності від розміру зображення

```python
metrics = [[],[],[],[],[]]
img_dir = "../images/AS_dataset/test"
data_path = os.path.join(img_dir,'*g') 
files = glob.glob(data_path) 

sift = cv.SIFT_create()

main_img = cv.imread("../images/AS_dataset/train/best_img.jpg", cv.IMREAD_GRAYSCALE)
# main_img = cv.normalize(main_img, None, 0, 255, cv.NORM_MINMAX).astype('uint8')
kp1, des1 = sift.detectAndCompute(main_img,None)

# print('Start ' + str(i) + 'dataset')
for f1 in files: 
    img = cv.imread(f1,cv.IMREAD_GRAYSCALE) # trainImage
    img = cv.normalize(img, None, 0, 255, cv.NORM_MINMAX).astype('uint8')
    filename = f1.replace("../images/AS_dataset/test\\",'')

    #downscale image
    scale_percent = 80 # percent of original size
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)
    dim = (width, height)
    img = cv.resize(img, dim, interpolation = cv.INTER_AREA)


    e1 = cv.getTickCount()
    kp2, des2 = sift.detectAndCompute(img,None)
    e2 = cv.getTickCount()
    time = (e2 - e1)/ cv.getTickFrequency()
    img_h, img_w = img.shape


    if len(kp2)*len(kp1) > 0:
        bf = cv.BFMatcher()
        matches = bf.knnMatch(des1, des2, k=2)
        good = []
        distance_sum = 0
        for m,n in matches:
            if m.distance < 0.75 * n.distance:
                good.append([m])
                distance_sum += m.distance

        right_features_ratio = len(good)/len(matches)
        if len(good) == 0:
            avg_distance = 0
        else:
            avg_distance = distance_sum / len(good)
    else:
        right_features_ratio = -1
        avg_distance = -1
    img_size = img_h * img_w

    print('1st metric  right_features_ratio = ' + str(right_features_ratio))
    print('2nd metric  avg_distance= ' + str(avg_distance))
    print('3rd metric  time = ' + str(time))


#     img3 = cv.drawMatchesKnn(main_img,kp1,img,kp2,good,None,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
#     plt.imshow(img3),plt.show()
    metrics[0].append(filename)
    metrics[1].append(right_features_ratio)
    metrics[2].append(avg_distance)
    metrics[3].append(time)
    metrics[4].append(img_size)
        
df = pd.DataFrame({"filename " : metrics[0], "metric1 ": metrics[1], "metric2": metrics[2], "metric3 ": metrics[3], "pixels ": metrics[4]})        
df.to_csv('AS_metrics.csv')
```
